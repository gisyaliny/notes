### Overview of NFIP Requirements

- **NFIP**: National Flood Insurance Program.
- **CFR**: Code of Federal Regulations, specifically **44 CFR** is crucial for understanding NFIP.
- **Subsection 60-70**: Detailed requirements found in **24 CFR**.


### Step 1: Handling Areas Without Maps

- **60.3(a)**: Communities without maps must have officials determine if areas are reasonably safe from flooding.
  * Responsibility falls on local officials (building officials, floodplain administrators, mayors).
  * Challenges: Balancing safety and community development costs.


### Responsibilities and Regulations

#### Clean Water Act
- **Enforcement**: U.S. Army Corps of Engineers (USACE).
- **Importance**: Regulates construction near navigable waters.

#### Executive Order 11988 (1986)
- **Purpose**: Requires federal agencies to assess if projects are in the 100 or 500-year floodplain and explore alternative sites.
- **Example**: Building federal projects (e.g., libraries, nuclear power plants) outside floodplains is crucial due to potential flooding impacts.

    | Project Type | Floodplain Considerations |
    |-------------|---------------------------|
    | Federal Prisons | Floodplain concerns may arise. |
    | Federal Libraries | Must avoid floodplains to protect investments. |
    | Nuclear Power Plants | Critical to avoid floodplain locations. |


### Step 2: Base Flood Elevation (BFE) Requirements

- **Permit Requirement**: Any development in an approximate zone requires a permit.
- **Base Flood Elevation Data**: Developers must provide this if no maps are available.
  * **Thresholds**:
    * 50 Lots or 5 acres, whichever is less.

#### Example of Lot Considerations

| Scenario | Requirement |
|----------|-------------|
| 51 Lots, 4 acres | Requires BFE data. |
| 49 Lots, 6 acres | Requires BFE data. |
| 50 Lots, 5 acres | Requires BFE data. |

### Records and Documentation

- **Elevation Certificates**: Required to document the lowest floor elevation of structures.
  * **Retention**: Must be kept indefinitely with no statute of limitations.
  * In A zones, 1st floor is TOP of bottom floor.
  *   In V zones , 1st floor is BOTTOM of lowest horizontal  structure member
  
  

### Additional Considerations

- **Manufactured Homes**: Must adhere to specific NFIP standards.
- **Venting Requirements**: For areas beneath the lowest floor.
  * **Hydraulic Ports**:
    * Minimum of two vents required.
    * Area: 1 square inch for every square foot of enclosed space.

  | Area Size | Venting Requirement |
  |-----------|---------------------|
  | 1,000 square feet | 1,000 square inches of venting needed. |
  | 500 square feet | 500 square inches of venting needed. |


### Common Issues and Misunderstandings

- Misconceptions about venting and elevation requirements can lead to violations.
- **Example**: Incorrect installations (plywood blocking vents) may not meet NFIP standards.

### Key Takeaways

- Always ensure projects adhere to NFIP guidelines and standards.
- Local officials play a critical role in floodplain management, particularly in areas lacking maps.
- Understanding and documentation of flood elevation and venting requirements are essential for compliance and safety.


## ðŸŒŠ Floodplain Management and Regulations

### Introduction to Floodplain Management

"Floodplain management involves regulating land use in areas that are prone to flooding, ensuring safety and compliance with federal and local regulations."

### Key Terms and Concepts

- **Flood Hazard Areas**: Regions identified as being at risk for flooding, often designated by specific zones (e.g., A, B, V zones).
- **Recreational Vehicles (RVs)**: Vehicles that can be parked in flood hazard areas under certain conditions, specifically:
  * Must be on-site for fewer than **180 consecutive days**.
  * Must be **fully licensed** and ready for highway use.

### RV Regulations in Flood Hazard Areas

- **Conditions for RVs**:
  * If parked long-term, they must comply with the permit requirements for permanent construction.
  * RVs need to be roadworthy; if not, citations may be issued.

#### Example Scenario

"If an RV has a dead battery, a citation may not be issued. However, if it is not roadworthy (e.g., wheels off, skirting around it), a citation is likely."

### Flood Zones and Construction Requirements

#### Designation of Flood Zones

| Zone Type | Description |
|-----------|-------------|
| A Zone | Areas subject to flooding; requires specific construction standards. |
| B Zone | Areas with lesser flooding risk; different standards apply. |
| V Zone | Coastal areas with high-velocity flood risks; must meet stringent regulations. |


#### Key Construction Requirements

- **Elevation Standards**:
  * Structures in **V zones** must have the bottom of the lowest horizontal structural member at or above the Base Flood Elevation (BFE).
  * **New construction** must be landward of high tide.

- **Engineering Certification**:
  * Required for construction in B zones, it verifies compliance with flood regulations.
  * Must be signed and sealed by a licensed engineer or architect.

### Building Considerations in Different Zones

#### Foundation Types and Measurements

- **Lowest Horizontal Structural Member**:
  * Important for determining compliance with flood elevation requirements.

- **Measurements**:
  * In A zones, measure from the top of the finished floor slab.
  * In V zones, measure from the bottom of the concrete slab.

#### Example of Measurement

"In the V zone, the bottom of the slab is measured for compliance, even if it seems counterintuitive."

### Elevation Certificates

#### Importance of Elevation Certificates

"Elevation certificates provide formal documentation of compliance with floodplain regulations."

- **Who Completes the Certificate**:
  * Typically completed by land surveyors, engineers, or architects.
  * Community officials rarely sign these due to potential conflicts of interest.

#### Considerations When Completing Elevation Certificates

| Scenario | Action |
|----------|--------|
| Surveyor can't access crawlspace | Use alternative documentation or contact the floodplain administrator. |
| Required data fields | Must include accurate elevation measurements and flood zone designation. |


### Definition and Handling of Basements

#### Understanding Basements

"A basement is defined as a floor level that is below ground on all sides, regardless of its use."

#### Important Considerations

- **Living Spaces**:
  * Even if used for living purposes, if a floor is below grade on all sides, it is classified as a basement.

- **Examples of Basements**:
  * A walk-out basement on three sides is still considered a basement.
  * Sunken living rooms that are below grade on all sides are also classified as basements.

### Record Retention and Compliance

#### Documentation Requirements

- **Retention of Elevation Certificates**:
  * No statute of limitations; must be stored securely for compliance.

- **Archival of Maps**:
  * Important for historical reference and compliance; communities should keep old maps for future determinations.

#### Benefits of Electronic Storage

"While not required, storing documents electronically can earn points in the Community Rating System (CRS) for floodplain management."

#### Conclusion on Record-Keeping

- Retaining accurate records is crucial for compliance and historical reference, particularly when assessing structures built many years ago.

### Overview of Section 60.3

Section 60.3 of the NFIP regulations outlines the specific floodplain management criteria that communities must adopt to participate in the NFIP. The requirements are progressive and increase based on the type of flood hazard data provided to the community.

### 60.3(a) â€” Communities Without FIRMs or FHBMs

- Communities must require permits for all proposed construction or development.
- Must review permit applications to determine if the proposed construction will be reasonably safe from flooding.
- Utilize any available data from federal, state, or other sources to identify flood-prone areas.
- Ensure all new construction is designed and adequately anchored to prevent flotation, collapse, or lateral movement.
- Use flood-resistant materials and methods.

### 60.3(b) â€” Communities With FHBMs But No FIRMs

- All requirements of 60.3(a) apply.
- Requires development permits in SFHA areas.
- Verification that encroachments don't increase flood levels during a 100-year flood.
- New construction must be elevated to or above the BFE.
- Prohibits encroachments in floodways if they would increase flood levels.

### 60.3(c) â€” Communities With FIRMs and BFEs

- All requirements of 60.3(a) and (b) apply.
- New construction must have lowest floor elevated to or above the BFE.
- Manufactured homes must be installed using methods that minimize flood damage.
- Non-residential structures can be floodproofed instead of elevated if certified by an engineer.
- Prohibits new construction in floodways that would increase base flood elevations.

### 60.3(d) â€” Communities With FIRMs, BFEs, and Floodways

- All requirements of 60.3(a), (b), and (c) apply.
- No encroachments in floodways unless certified by an engineer that there will be no increase in flood levels.
- Communities must obtain and review detailed hydraulic data for all proposed developments.
- Must notify adjacent communities and state coordinating agencies before altering watercourses.

### 60.3(e) â€” Communities With Coastal High Hazard Areas (V Zones)

- All requirements of 60.3(a), (b), (c), and (d) apply.
- New construction must be located landward of the mean high tide line.
- Structures must be elevated on pilings/columns and have space beneath the lowest floor free of obstructions.
- No use of fill for structural support.
- No alteration of sand dunes or mangrove stands that would increase flood damage.

### Key Enforcement Points for Floodplain Administrators

- **Substantial Improvement/Damage**: Any reconstruction, rehabilitation, addition, or other improvement of a structure where the cost equals or exceeds 50% of the market value of the structure before the improvement or damage occurred.
- **Cumulative Substantial Improvement**: Some communities track cumulative improvements over time. If they collectively exceed 50% of value, the structure must be brought into compliance.
- **Freeboard**: Additional height (typically 1-3 feet) above the BFE required by many communities for safety margin.
- **Critical Facilities**: Special treatment for facilities like hospitals, fire stations, emergency operations centers, etc. Often required to be elevated higher than standard structures.

### Common Compliance Issues

- **Enclosure Violations**: Using areas below BFE for purposes other than parking, storage, or building access.
- **Venting Problems**: Insufficient flood vents or improper installation of vents.
- **Elevation Certificate Errors**: Inaccurate measurements or misidentification of the lowest floor.
- **Improper Wet Floodproofing**: Using inappropriate materials or techniques below the BFE.
- **Fill Issues**: Improper placement or inadequate compaction of fill material.

### Community Rating System (CRS) Connection to Section 60.3

Communities that adopt floodplain management standards that exceed the minimum requirements of Section 60.3 may qualify for reduced flood insurance premiums through the CRS. Some credited activities include:

- Adopting higher freeboard requirements
- Implementing stricter standards for critical facilities
- Requiring compensatory storage for fill in the floodplain
- More restrictive floodway encroachment standards
- Protection of floodplain functions and natural areas
- Higher substantial damage/improvement thresholds

### Documentation Requirements for Section 60.3 Compliance

- **Building Permits**: Must document elevation requirements and construction standards.
- **Variance Procedures**: Must have clear criteria and review process for variance requests.
- **As-Built Certifications**: Required to verify that construction meets approved plans.
- **Inspection Records**: Document site visits during construction phases.
- **Violation Notices**: Formal documentation of non-compliance and remedial actions.
- **Floodplain Development Permits**: Separate from or integrated with building permits.
