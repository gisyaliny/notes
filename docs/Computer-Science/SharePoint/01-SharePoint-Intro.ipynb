## What is SharePoint?

- [Reference](https://www.youtube.com/watch?v=Mu6oWtbk9Nc&list=PLoyECfvEFOjZ5avY1R9dcSZ3ZKGmuk1O4&ab_channel=LearnitTraining)

- SharePoint is a web-based collaboration platform that integrates with Microsoft Office. It enables teams to store, organize, share, and access information from any device with an internet connection.
- SharePoint can be used for various purposes such as `document management`, `team collaboration`, `intranet portals`, `project management`, and more.


## SharePoint Site Structure

- `Sub Sites`: Works for different department
  - Assign people/role with different permission to access these sites




## Getting Started

- [Reference](https://learn.microsoft.com/en-us/sharepoint/add-sample-site)

1. Browse the samples in the online [SharePoint look book](https://lookbook.microsoft.com/), Select a sample and `Add to your tenant`.




2. Enter your information, verify it
   Next, enter your email address (for a notification of when your site is ready to use), the title you want to use for your site, and the site URL you want to use.

3. Deploy your selected sample
   1. Select Provision, and in a short time your site will be ready for you to use.
